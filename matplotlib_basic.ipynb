# Matploblib 라이브러리의 기초
Matplotlib란?
- 다양한 데이터를 시각화 할 수 있도록 도와주는 라이브러리.
- 간단한 데이터 분석에서부터 인공지능 모델의 시각화까지 활용도가 매우 높음.

```python
# 긴단한 직선 그래프 그리기
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [1, 2, 3]
plt.plot(x, y)
plt.title("linear plot")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
plt.savefig("picture.png") # save at local computer
```

```python
# 그래프 저장하기
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, np.pi*10, 500) # pi * 10 너비에, 500개의 점들을 균일하게 찍기
fig, axes = plt.subplots(2, 1) # 2개의 그래프가 들어가는 figure 생성
axes[0].plot(x, np.sin(x))
axes[1].plot(x, np.cos(x))
plt.show()
# fig.savefig("sin&cos.png")
```

```python
# 선 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9, 10)
y = x ** 2
# 라인 스타일로는 "-", ":", "-.", "--" 등이 사용될 수 있다.
plt.plot(x, y, linestyle=":", marker="*")
# x, y축에서 특정 범위를 자를 수 있음.
plt.show()
```

```python
# 선 그래프 그리기
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-9, 10)
y1 = x ** 2
y2 = -x
plt.plot(x, y1, linestyle="-", marker="*", color="red", label="y = x * x")
plt.plot(x, y2, linestyle=":", marker="o", color="blue", label="y = -x")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend(
    shadow=True,
    borderpad = 1
) # label 박스에 넣기
plt.show()
```
